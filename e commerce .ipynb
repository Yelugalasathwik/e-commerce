```python
#Appending producct url to a list
col=[]
for i, row in meesho.iterrows():
  col.append(row[2])
```

```python
#Fetching product name from product url
prodnames=[]
pattern = r"/([^/]+)/p/"
for i in range(len(col)):
    match = re.search(pattern, col[i])
    if match:
        extracted_string =match.group(1)
        prodnames.append(extracted_string)
```

```python
meesho['product_name']=prodnames
```

## top 50

```python
#using popularity index to get top 50 products
num_rating =  meesho.groupby('product_name').count()['rating'].reset_index()
num_rating.rename(columns={'rating':'num_ratings'},inplace=True)
num_rating
```

```python
avg_rating =  meesho.groupby('product_name').mean()['rating'].reset_index()
avg_rating.rename(columns={'rating':'avg_ratings'},inplace=True)
avg_rating
```

```python
pop_50 = num_rating.merge(avg_rating,on='product_name')
pop_50
```

```python
pop_50[pop_50['num_ratings']>=300]
```

```python
#Displaying popular products 
pop_50_sorted = pop_50[pop_50['num_ratings'] >= 300].sort_values(by='avg_ratings',ascending = False).head(50)
pop_50_sorted['avg_ratings'] = round(pop_50_sorted['avg_ratings'], 2)
pop_50_products = pop_50_sorted.merge(meesho,on='product_name').drop_duplicates('product_name')[['product_name','product_id','image_url','price','num_ratings','avg_ratings']]
pop_50_products
```

```python
col=[]
for i, row in meesho.iterrows():
    col.append(row[2])
```

```python
prodnames=[]

pattern = r"/([^/]+)/p/"
for i in range(len(col)):
    match = re.search(pattern, col[i])
    if match:
        extracted_string =match.group(1)
        prodnames.append(extracted_string)
```

```python
meesho['product_name']=prodnames
```

```python
#setting threshold for users who've rated more than 15 products
x=meesho.groupby('user_id').count()['rating'] > 15
valuable_user = x[x].index
valuable_user
```

```python
filtered_user_ratings = meesho[meesho['user_id'].isin(valuable_user)]
```

```python
filtered_user_ratings
```

```python
#setting threshhold for product names having more than 300 ratings
y = filtered_user_ratings.groupby('product_name').count()['rating'] >= 300
popularProductname=y[y].index
popularProductname
```

```python
final_ratings_name = filtered_user_ratings[filtered_user_ratings['product_name'].isin(popularProductname)]
```

```python
final_ratings_name.drop_duplicates()
```

```python
#create pivot table from df for product name
pivotByname=final_ratings_name.pivot_table(index='product_name',columns='user_id',values='rating')
```

```python
pivotByname
```

```python
pivotByname.fillna(0,inplace=True)
pivotByname
```

```python
similarity_score_by_name = cosine_similarity(pivotByname)
similarity_score_by_name
```

```python
similarity_score_by_name.shape
```

```python
result = pd.DataFrame(similarity_score_by_name)
result.to_csv("Similarity_matrix.csv" , index=False)
```

```python
# function to display 5 products by product name for web page

import numpy as np
def recommendByProductname(prod_name):
    #fetch index
    index=np.where(pivotByname.index == prod_name)[0][0]
    #print(index)
    similar_items=sorted(list(enumerate(similarity_score_by_name[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    for i in similar_items:
        item = []
        temp_df = meesho[meesho['product_name']==pivotByname.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('product_name')['image_url'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['price'].values))
        item.extend(list(temp_df.drop_duplicates('product_name')['rating'].values))
        data.append(item)
    return data
```
